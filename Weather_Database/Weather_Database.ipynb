```python
#import dependencies
import numpy as np
import requests
import json
import pandas as pd

#importing citipy to get nearest city
from citipy import citipy

# Import API key
from config import api_key
```

```python
#generating random floats for lat and lng 
lats = np.random.uniform(-90,90,2000)
lngs = np.random.uniform(-180,180,2000)

#creating lat and lng combinations
lat_lngs = []
lat_lngs = zip(lats,lngs)
```

```python
#creating empty array to append the name of cities
cities=[]

#iterating through combinations of lat and lng to find the nearest city
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name
    
    #if the city value is unique, add to cities[]
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

```python

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# initialising empty columns in df to hold reponse info     
cities_df = pd.DataFrame({})

cities_df["city"] = cities             
cities_df["Country"] = ""
cities_df["Lat"] =""
cities_df["Lng"]=""
cities_df["Max Temp"]=""
cities_df["Humidity"]=""
cities_df["Cloudiness"]=""
cities_df["Wind Speed"]=""
cities_df["Current Description"]=""




#calling API
base_url = "http://api.openweathermap.org/data/2.5/weather?appid=" + api_key

```

```python
# Loop through all the cities in our list.
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = base_url + "&q=" + city.replace(" ","+")
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
    
    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        cities_df["Country"][i] = city_weather["sys"]["country"]
        cities_df["Lat"][i] = city_weather["coord"]["lat"]
        cities_df["Lng"][i]= city_weather["coord"]["lon"]
        cities_df["Max Temp"][i]= city_weather["main"]["temp_max"]
        cities_df["Humidity"][i]= city_weather["main"]["humidity"]
        cities_df["Cloudiness"][i]= city_weather["clouds"]["all"]
        cities_df["Wind Speed"][i]= city_weather["wind"]["speed"]
        cities_df["Current Description"][i] = city_weather["weather"][0]["description"]
        

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```
